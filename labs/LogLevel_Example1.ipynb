# Example 1. Salary and Education



#### To study the relationship between education and salaries, we postulate the simple model:

$ log(salary)=\beta_0+\beta_1*educ+u $



### Using the data in WAGE1 where n=526 individuals

```python
salaries = woo.dataWoo('wage1')
```

```python
salaries.head()
```

```python
salaries.describe()
```

```python
type(salaries)
```

```python
# We impose a simple, linear, model: 
# We specify CeoSalaries as the empirical dataset

reg = smf.ols(formula='np.log(wage) ~ educ', data=salaries)
```

```python
# We fit the model
results = reg.fit()
```

```python
b = results.params
print(f'b: \n{b}\n')
```

## Based on the previous we have fitted the following model:

$ log(wage)=0.584+0.083*educ+u $


## How do we interpret the equation?

#### The coefficient on educ has a percentage interpretation when it is multiplied by 100: salary increases by 8.3% for every additional year of education

```python
sns.scatterplot(data=salaries, x='educ',y='wage')


educationValues=[item for item in range(0,18,1)]

yHat=[np.exp(0.584+0.083*item) for item in range(0,18,1)]

sns.lineplot(educationValues, yHat,color='purple')
```

#### We observe from the previous graph that the model is non-linear as expected given the non-linear component we specified
